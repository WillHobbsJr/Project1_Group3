```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame
data = {
    'Name': ['Metformin', 'Phenformin', 'Mounjaro'],
    'Side Effects ': ['Breathing Problems, Lactic Acidosis', 'Dizziness, Vomiting', 'Thyroid Cancer']
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)  # Using print to show the DataFrame in the output

# Splitting the 'Side Effects' column into individual effects
df['Side Effects'] = df['Side Effects '].str.split(',')

# Counting the number of side effects for each drug
df['Number of Side Effects'] = df['Side Effects'].apply(len)

# Creating a bar plot
plt.bar(df['Name'], df['Number of Side Effects'])
plt.xlabel('Drug Name')
plt.ylabel('Number of Side Effects')
plt.title('Number of Side Effects for Different Drugs')
plt.show()
```
